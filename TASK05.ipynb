```python
plt.figure(figsize=(8,6))
sns.countplot(x="WeatherCondition",data=df,order=df['WeatherCondition'].value_counts().index)
plt.title("Distribution of Accident based on Weather Conditions")
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,6))
sns.countplot(x="RoadCondition",hue="Timing",data=df,order=df['RoadCondition'].value_counts().index)
plt.title("Accidents by RoadCondition and Timing of the Day")
plt.xticks(rotation=45,ha="right")
plt.show()

```

```python
plt.figure(figsize=(8,6))
state_accidents=df['State'].value_counts()
sns.barplot(x=state_accidents.index,y=state_accidents.values,palette="viridis")
plt.title("Accident Hotspots by States")
plt.xlabel("State")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45,ha="right")
plt.show()

```

```python
plt.figure(figsize=(8,6))
reasons_accidents=df['Reason'].value_counts()
sns.barplot(x=reasons_accidents.index,y=reasons_accidents.values,palette="muted")
plt.title("Distribution of Accidents Based on Reasons")
plt.xlabel("Reason")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45,ha="right")
plt.show()
```
